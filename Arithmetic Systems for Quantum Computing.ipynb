# Fundamental Arithmetic Operations with Quantum Computers
## Necessary Imports

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, Aer, execute
from qiskit.visualization import plot_state_city,plot_state_qsphere
```

## Creating Backend

```python
be=Aer.get_backend("statevector_simulator")
```

## 1-Bit Multiplication

```python
in_=QuantumRegister(2,"in")
out_=QuantumRegister(1,"out")
qc=QuantumCircuit(in_,out_)

qc.h(0)
qc.h(1)
qc.ccx(0,1,2)

display(qc.draw("mpl"))
display(plot_state_qsphere(execute(qc,be).result().get_statevector()))
```

## Half Adder

```python
in_=QuantumRegister(2,"in")
out=QuantumRegister(1,"out")
carry=QuantumRegister(1,"carry")
qc=QuantumCircuit(in_,out,carry)

qc.h(0)
qc.h(1)
qc.cx(0,2)
qc.cx(1,2)

qc.ccx(0,1,3)

display(qc.draw("mpl"))
display(plot_state_qsphere(execute(qc,be).result().get_statevector()))
```

## Full Adder

```python
in_=QuantumRegister(2,"in")
c_in=QuantumRegister(1,"c-in")
out=QuantumRegister(1,"out")
c_out=QuantumRegister(1,"c-out")
t=QuantumRegister(3,"tmp")
qc=QuantumCircuit(in_,c_in,t,c_out,out)

qc.h(0)
qc.h(1)
qc.h(2)
qc.cx(0,3)
qc.cx(1,3)
qc.cx(3,7)
qc.cx(2,7)
qc.ccx(0,1,4)
qc.ccx(2,3,5)
qc.cx(5,6)
qc.cx(4,6)
qc.ccx(5,4,6)

display(qc.draw("mpl"))
display(plot_state_qsphere(execute(qc,be).result().get_statevector()))
```

## Half Subtractor

```python
in_=QuantumRegister(2,"in")
b=QuantumRegister(1,"b")
out=QuantumRegister(1,"out")
qc=QuantumCircuit(in_,b,out)

qc.h(0)
qc.h(1)

qc.cx(0,3)
qc.cx(1,3)
qc.ccx(1,3,2)

display(qc.draw("mpl"))
display(plot_state_qsphere(execute(qc,be).result().get_statevector()))
```

## Full Subtractor

```python
in_=QuantumRegister(2,"in")
b_in=QuantumRegister(1,"b-in")
b_out=QuantumRegister(1,"b-out")
tmp=QuantumRegister(3,"t")
out=QuantumRegister(1,"difference")
qc=QuantumCircuit(in_,b_in,tmp,b_out,out)

qc.h(0)
qc.h(1)
qc.h(2)

qc.cx(0,3)
qc.cx(1,3)

qc.cx(3,7)
qc.cx(2,7)

qc.x(0)
qc.ccx(0,1,4)
qc.x(3)
qc.ccx(2,3,5)

qc.cx(4,6)
qc.cx(5,6)
qc.ccx(4,5,6)

display(qc.draw("mpl"))
display(plot_state_qsphere(execute(qc,be).result().get_statevector()))
```
